```python
import pandas as pd
import yfinance as yf
import datetime
from datetime import datetime
import numpy as np
```

```python
Favored_Basket = pd.read_excel("C:/Users/Leon/Baskets.xlsx", 'Favored Basket')
Disfavored_Basket = pd.read_excel("C:/Users/Leon/Baskets.xlsx", 'Disfavored basket')
```

```python
Favored_Basket_tickers_list = Favored_Basket['Ticker'].to_list()
Disfavored_Basket_tickers_list = Disfavored_Basket['Ticker'].to_list()
```

```python
start = datetime(2022,11,21)
now = datetime.now().date()
```

```python
favored_data = yf.download(Favored_Basket_tickers_list, start=start, end=now)
disfavored_data = yf.download(Disfavored_Basket_tickers_list, start=start, end=now)
```

```python
now = np.datetime64(favored_data.tail(1).index[0])
```

```python
favored_data = favored_data['Adj Close'].transpose()
disfavored_data = disfavored_data['Adj Close'].transpose()
```

```python
first_date = np.datetime64('2022-11-21')
```

```python
favored_data['return']=((favored_data[now]-favored_data[first_date])*100)/favored_data[first_date]
disfavored_data['return']=((disfavored_data[now]-disfavored_data[first_date])*100)/disfavored_data[first_date]
```

```python
favored_data
```

```python
disfavored_data
```

```python
faveored_basket_mean = favored_data['return'].mean()
disfavored_basket_mean = disfavored_data['return'].mean()
print(f"Favored Basket mean from {first_date} until {now.astype(datetime).date()} is: {round(faveored_basket_mean,4)}")
print(f"Disfavored Basket mean from {first_date} until {now.astype(datetime).date()} is: {round(disfavored_basket_mean,4)}")
```

```python
all_data = pd.concat([favored_data,disfavored_data])
```

```python
sorted_data=all_data['return'].sort_values(ascending=False)
```

```python
sorted_data = pd.DataFrame(sorted_data)
```

```python
sorted_data
```

```python
favored_tickers_list = list(favored_data.index)
sorted_ticker_list = list(sorted_data.index)
```

```python
disfavored_tickers_list = list(disfavored_data.index)
```

```python
top_48 = sorted_ticker_list[:48]
```

```python
last_54 = sorted_ticker_list[48:]
```

```python
favored_dict_score = {}
for ticker in favored_tickers_list:
    if ticker in top_48:
        favored_dict_score[ticker] = 1
    else:
        favored_dict_score[ticker] = -1
```

```python
disfavored_dict_score = {}
for ticker in disfavored_tickers_list :
    if ticker in last_54:
        disfavored_dict_score[ticker] = 1
    else:
        disfavored_dict_score[ticker] = -1
```

```python
favored_sum = sum(favored_dict_score.values())
```

```python
disfavored_sum = sum(disfavored_dict_score.values())
```

```python
print(f"The favored basket sum: {favored_sum}, The disfavored basket sum:{disfavored_sum}")
```

# It means that in favored basket the score is close to 50/50. in 26 out of 48 the analysis was right. For the disfavored basket the score is a bit better and stands at 35 correct out of 54.
